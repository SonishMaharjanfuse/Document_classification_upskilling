# Training

```python

# import YOLO model
from ultralytics import YOLO

# Load a model
model = YOLO('yolov8n-cls.pt')  # load a pretrained model

# Train the model
model.train(data='/content/data', epochs=5)
```

# Save the model

```python
# Save the model
model.export()
```

# Evaluate the model

```python
# Validate the model
metrics = model.val()  # no arguments needed, dataset and settings remembered
metrics.top1  # top1 accuracy
metrics.top5  # top5 accuracy`
```

# Inference

```python
# Load the best model for inference
model2 = YOLO('/content/runs/classify/train/weights/best.pt')
```

```python
import numpy as np
from PIL import Image
image_path = "/content/test2.jpeg"
Image.open(image_path)
results = model2.predict(image_path)
name_dic = results[0].names
key = np.argmax(results[0].probs.data).item()
print(f"prediction:{name_dic[key]}")
```
